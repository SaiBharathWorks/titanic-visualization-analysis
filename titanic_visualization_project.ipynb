```python
# TITANIC DATA VISUALIZATION PROJECT

import matplotlib.pyplot as plt
import seaborn as sns

# 1. Load Titanic Dataset
titanic = sns.load_dataset('titanic')
print(titanic.head())

# 2. Bar Chart – Survival Count
survival_counts = titanic['survived'].value_counts()
plt.bar(['Not Survived', 'Survived'], survival_counts, color=['red', 'green'])
plt.xlabel("Status")
plt.ylabel("Number of People")
plt.title("Titanic Survival Count")
plt.show()

# 3. Bar Chart – Gender-wise Survival
gender_survival = titanic.groupby('sex')['survived'].sum()
plt.bar(gender_survival.index, gender_survival.values, color=['blue', 'pink'])
plt.xlabel("Gender")
plt.ylabel("Survived (Count)")
plt.title("Titanic Survival by Gender")
plt.show()

# 4. Scatter Plot – Age vs Fare
titanic_clean = titanic.dropna(subset=['age', 'fare'])
plt.scatter(titanic_clean['age'], titanic_clean['fare'], color='purple', alpha=0.5)
plt.xlabel("Age")
plt.ylabel("Fare Paid ($)")
plt.title("Age vs Fare Paid (Titanic Passengers)")
plt.show()

# 5. Histogram – Age Distribution
ages = titanic.dropna(subset=['age'])['age']
plt.hist(ages, bins=10, color='skyblue', edgecolor='black')
plt.xlabel("Age Groups")
plt.ylabel("Number of Passengers")
plt.title("Age Distribution of Titanic Passengers")
plt.show()
```
